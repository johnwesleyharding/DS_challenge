```python
import pandas as pd
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('puzzle.csv', header = None)
```

```python
df.columns = ['X', 'y']
print(df.shape)
df.head()
```

```python
df.plot.scatter(x = 'X', y = 'y')
```

```python
for n in range(2, 11):
    
    model = KMeans(n_clusters = n)
    model.fit(df)
    df.plot.scatter(x = 'X', y = 'y', c = model.labels_, cmap = 'rainbow')
```
