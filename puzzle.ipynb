```python
import pandas as pd
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('puzzle.csv', header = None)
```

```python
df.plot.scatter(x = 0, y = 1)
```

```python
sns.jointplot(x = 0, y = 1, data = df, kind = 'hex')
```

```python
for n in range(2, 11):
    
    model = KMeans(n_clusters = n)
    model.fit(df)
    df.plot.scatter(x = 0, y = 1, c = model.labels_, cmap = 'rainbow')
```
